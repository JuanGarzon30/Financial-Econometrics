El siguiente programa tiene como objetivo valorar opciones, instrumentos derivados mediante los cuales el comprador del mismo
tiene la potestad de comprar o vender cierto activo a un precio determinado en una fecha específica. Para realizarlo, se tuvo en cuenta el modelo de Black-Scholes, reconocido como uno de los más sencillos y rápidos para la valoración de opciones. Así mismo, muestra las gráficas de 5 letras griegas, mediante las cuales el usuario podrá ver la sensibilidad que tienen las opciones con respecto a variables como los cambios en el precio spot, la volatilidad, el tiempo y la tasa de interés.

Este programa pide al usuario ciertos inputs. Dado que es un programa creado para valorar opciones sobre divisas, el programa le pide al usuario la divisa con la cual quiere averiguar el precio de las opciones (con respecto al dólar estadounidense). Así mismo, da al usuario la opción de que elija qué griega quiere ver graficada.

Por otra parte, pide las condiciones básicas para el desarrollo del modelo, tales como:

1. El precio strike, o de ejercicio, el cual es acordado por el comprador de la opción para saber a qué precio deseará comprar o vender el activo subyacente.

2. La tasa de interés.

3. El tipo de opción: call (de compra), o put (de venta).

4. Los días que mantendrá la opción.

Estas cuatro variables no son suficientes para la realización del modelo. Sin embargo, lo novedoso de este programa es que toma información en tiempo real y actualizada sobre el comportamiento de la divisa elegida en el mercado a través de la API de AlphaVantage. A través de esta, se desarrollan las tablas en pandas que permitirán la realización del programa en su totalidad, recopilando información sobre el precio Spot del activo y su volatilidad histórica, con más de 2000 filas de información.

Este programa es altamente valioso por dos motivos:

1. Toma información en tiempo real de las 157 divisas que maneja la API, y usa la información de manera rápida para llevar a cabo los cálculos necesarios y mostrar los resultados en forma de precio y gráficas.

2. El manejo de opciones en el mundo, aunque bastante importante, es desconocido por muchos. De esta forma, tener una herramienta que permita visualizar de forma veloz los precios justos de estos instrumentos derivados y las afectaciones que pueden tener variables como el tiempo, la tasa de interés, el spot y la volatilidad es de suma importancia y utilidad al momento de operar con estos activos financieros.

```python
from tkinter import *
from tkinter import ttk
import tkinter as tk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statistics import NormalDist
from alpha_vantage.foreignexchange import ForeignExchange
key = "L78L88TMX4SKPBIN"

### variables ###

opción = 0
spot = 0
strike = 0
tasa = 0
dias = 0
años = dias / 360
volatilidad = 0
 
n_d1 = []
n_prima_d1 = []
n_neg_d1 = []
n_d2 = []
n_neg_d2 = []    
   
### Enunciado ###
    
ventana = Tk()
ventana.title("Black-Scholes Model")
ventana.geometry("1200x1000")
ventana.resizable(False,False)
ventana.config(background="#4391d7")

etiqueta=tk.Label(ventana, text= "Modelo Black-Scholes", font= ("Arial Black", 30), background="#4391d7")
etiqueta.pack()
etiqueta.place(x=370, y=60)

etiqueta=tk.Label(ventana, text= "Este modelo puede ser utilizado para valorar opciones y graficar las distintas sensibilidades de los\n instrumentos derivados.", font=("Arial", 18), background="#4391d7")
etiqueta.pack()
etiqueta.place(x=50, y=155)

etiqueta=tk.Label(ventana, text= "A continuación seleccione la moneda que desea analizar: ", font=("Arial", 15), background="#4391d7")
etiqueta.pack()
etiqueta.place(x=100, y=280)

eleccionDivisas=ttk.Combobox(ventana, state="readonly", values = code)
eleccionDivisas.place(x=610, y=287)

etiqueta=tk.Label(ventana, text= "Por medio de que Griega desea analizar la sensibilidad: ", font=("Arial", 15), background="#4391d7")
etiqueta.pack()
etiqueta.place(x=100, y=350)

eleccionGriega=ttk.Combobox(ventana, state="readonly", values = ["Vega","Gamma","Delta","Theta","Rho"])
eleccionGriega.place(x=610, y=357)

etiqueta=tk.Label(ventana, text= "Digite las condiciones iniciales: ", font=("Bahnschrift SemiBold", 18), background="#4391d7")
etiqueta.pack()
etiqueta.place(x=100, y=435)

#### Inputs ####

etiquetaStrike=tk.Label(ventana, text= "Precio Strike", font=("Arial", 15), background="#4391d7")
etiquetaStrike.pack()
etiquetaStrike.place(x=120, y=500)

strikeprice=tk.Entry(ventana)

strikeprice.place(x=280, y=505)

etiquetaTipoOpcion=tk.Label(ventana, text= "Tipo de opción", font=("Arial", 15), background="#4391d7")
etiquetaTipoOpcion.pack()
etiquetaTipoOpcion.place(x=120, y=535)

eleccionTipoOpcion=ttk.Combobox(ventana, state="readonly", values = ["put","call"])
eleccionTipoOpcion.place(x=280, y=540)

etiquetaTInteres=tk.Label(ventana, text= "Tasa de interés", font=("Arial", 15), background="#4391d7")
etiquetaTInteres.pack()
etiquetaTInteres.place(x=120, y=570)

entradaTInteres=tk.Entry(ventana)
entradaTInteres.place(x=280, y=575)

porcentaje=tk.Label(ventana, text= "%", font=("Arial", 18), background="#4391d7")
porcentaje.place(x=410, y=570)

etiquetaDias=tk.Label(ventana, text= "Número de días", font=("Arial", 15), background="#4391d7")
etiquetaDias.pack()
etiquetaDias.place(x=120, y=605)

entradaDias=tk.Entry(ventana)

entradaDias.place(x=280, y=610)

#encabezados

encabezado_calculo=tk.Label(ventana, text= "El precio de la opción es:", font=("Arial", 15), background="#4391d7")
encabezado_calculo.place(x=700, y=436)

#botones

def black_scholes():

    opción = (eleccionTipoOpcion.get())
    strike = float(strikeprice.get())
    tasa = float(entradaTInteres.get())
    dias = float(entradaDias.get())
    años = dias / 360
    
    ts = ForeignExchange(key, output_format = "pandas")
    data_alphavantage, usdcop = ts.get_currency_exchange_daily("USD", eleccionDivisas.get(), "full")
    promedio = np.mean(data_alphavantage["4. close"])

    data_alphavantage["5. Variación Diaria"] = data_alphavantage["4. close"].pct_change(1) + 1
    data_alphavantage["6. Retorno Diario"] = np.log(data_alphavantage["5. Variación Diaria"])
    volatilidad = np.std(data_alphavantage["6. Retorno Diario"]) * np.sqrt(252)
    spot = data_alphavantage["4. close"][0]

    data_alphavantage

    #datos de opciones para griegas y calculo
    data_opciones = {"Spot": np.arange(strike * 0.1, strike * 3, 1)}

    df = pd.DataFrame(data_opciones)

    df["d1"] = (np.log(df["Spot"] / strike) + (tasa + ((volatilidad ** 2) / 2)) * años) / (volatilidad * np.sqrt(años))

    for d1 in df["d1"]:
        a = NormalDist().cdf(d1)
        n_d1.append(a)

    df["N(d1)"] = n_d1

    for d1 in df["d1"]:
        w = NormalDist().pdf(d1)
        n_prima_d1.append(w)

    df["N'(d1)"] = n_prima_d1

    df["-d1"] = - df["d1"]

    for neg_d1 in df["-d1"]:
        b = NormalDist().cdf(neg_d1)
        n_neg_d1.append(b)

    df["N(-d1)"] = n_neg_d1

    df["d2"] = (np.log(df["Spot"] / strike) + (tasa - ((volatilidad ** 2) / 2)) * años) / (volatilidad * np.sqrt(años))

    for d2 in df["d2"]:
        c = NormalDist().cdf(d2)
        n_d2.append(c)

    df["N(d2)"] = n_d2

    df["-d2"] = - df["d2"]

    for neg_d2 in df["-d2"]:
        d = NormalDist().cdf(neg_d2)
        n_neg_d2.append(d)

    df["N(-d2)"] = n_neg_d2

    df["Delta Call"] = df["N(d1)"]
    df["Delta Put"] = df["N(d1)"] - 1

    df["Theta Call"] = - ((df["Spot"] * df["N'(d1)"] * volatilidad) / (2 * np.sqrt(años))) - (tasa * strike * np.exp((- tasa) * años) * df["N(d2)"])
    df["Theta Put"] = - ((df["Spot"] * df["N'(d1)"] * volatilidad) / (2 * np.sqrt(años))) + (tasa * strike * np.exp((- tasa) * años) * df["N(-d2)"])

    df["Gamma"] = (df["N'(d1)"]) / (df["Spot"] * volatilidad * np.sqrt(años))

    df["Vega"] = df["Spot"] * np.sqrt(años) * df["N'(d1)"]

    df["Rho Call"] = strike * años * np.exp((- tasa) * años) * df["N(d2)"]
    df["Rho Put"] = - strike * años * np.exp((- tasa) * años) * df["N(-d2)"]

    print(df)
    print(data_alphavantage)

    spot = float(data_alphavantage["4. close"][0])
    volatilidad = np.std(data_alphavantage["6. Retorno Diario"]) * np.sqrt(252)

    print(spot)
    print(volatilidad)

    d1 = (np.log(spot / strike) + (tasa + ((volatilidad ** 2) / 2)) * años) / (volatilidad * np.sqrt(años))
    N_d1 = NormalDist().cdf(d1)
    N_neg_d1 = NormalDist().cdf(- d1)

    d2 = (np.log(spot / strike) + (tasa - ((volatilidad ** 2) / 2)) * años) / (volatilidad * np.sqrt(años))
    N_d2 = NormalDist().cdf(d2)
    N_neg_d2 = NormalDist().cdf(- d2)

    if opción == "call":
        prima = np.exp((- tasa) * años) * (spot * np.exp(tasa * años) * N_d1 - strike * N_d2)
        resultadoPrima = print("La prima de la opción tiene un valor de $" + str(prima))

    if opción == "put":
        prima = np.exp((- tasa) * años) * (strike * N_neg_d2 - spot * np.exp(tasa * años) * N_neg_d1)
        resultadoPrima = print("La prima de la opción tiene un valor de $" + str(prima))  

    labelRespuesta.config(text=f"El valor de la prima para la\n opción seleccionada es:\n ${str(prima)}", font =("Arial", 15)) 
        
    if eleccionGriega.get() == "Delta" and opción == "call":
        fig = Figure(figsize=(0.4, 0.4), dpi=100)
        t = np.arange(0, 3, .01)
        fig.add_subplot(111).plot(df["Delta Call"])

        canvas = FigureCanvasTkAgg(fig, master=ventana)  # A tk.DrawingArea.
        canvas.draw()
        canvas.get_tk_widget().place(x=400, y=690, width= 400, height= 230)
        
        toolbar = NavigationToolbar2Tk(canvas, root)
        toolbar.update()
        canvas.get_tk_widget().place(x=540, y=490, width= 150, height= 150)
        

    if eleccionGriega.get() == "Delta" and opción == "put":
        fig = Figure(figsize=(0.4, 0.4), dpi=100)
        t = np.arange(0, 3, .01)
        fig.add_subplot(111).plot(df["Delta Put"])

        canvas = FigureCanvasTkAgg(fig, master=ventana)  # A tk.DrawingArea.
        canvas.draw()
        canvas.get_tk_widget().place(x=400, y=690, width= 400, height= 230)
        
        toolbar = NavigationToolbar2Tk(canvas, root)
        toolbar.update()
        canvas.get_tk_widget().place(x=540, y=490, width= 150, height= 150)

    if eleccionGriega.get() == "Theta" and opción == "call":
        fig = Figure(figsize=(0.4, 0.4), dpi=100)
        t = np.arange(0, 3, .01)
        fig.add_subplot(111).plot(df["Theta Call"])

        canvas = FigureCanvasTkAgg(fig, master=ventana)  # A tk.DrawingArea.
        canvas.draw()
        canvas.get_tk_widget().place(x=400, y=690, width= 400, height= 230)
        
        toolbar = NavigationToolbar2Tk(canvas, root)
        toolbar.update()
        canvas.get_tk_widget().place(x=540, y=490, width= 150, height= 150)

    if eleccionGriega.get() == "Theta" and opción == "put":
        fig = Figure(figsize=(0.4, 0.4), dpi=100)
        t = np.arange(0, 3, .01)
        fig.add_subplot(111).plot(df["Theta Put"])

        canvas = FigureCanvasTkAgg(fig, master=ventana)  # A tk.DrawingArea.
        canvas.draw()
        canvas.get_tk_widget().place(x=400, y=690, width= 400, height= 230)
        
        toolbar = NavigationToolbar2Tk(canvas, root)
        toolbar.update()
        canvas.get_tk_widget().place(x=540, y=490, width= 150, height= 150)
    
    if eleccionGriega.get() == "Gamma":
        fig = Figure(figsize=(0.4, 0.4), dpi=100)
        t = np.arange(0, 3, .01)
        fig.add_subplot(111).plot(df["Gammal"])

        canvas = FigureCanvasTkAgg(fig, master=ventana)  # A tk.DrawingArea.
        canvas.draw()
        canvas.get_tk_widget().place(x=400, y=690, width= 400, height= 230)
        
        toolbar = NavigationToolbar2Tk(canvas, root)
        toolbar.update()
        canvas.get_tk_widget().place(x=540, y=490, width= 150, height= 150)
        
    if eleccionGriega.get() == "Vega":
        fig = Figure(figsize=(0.4, 0.4), dpi=100)
        t = np.arange(0, 3, .01)
        fig.add_subplot(111).plot(df["Vega"])

        canvas = FigureCanvasTkAgg(fig, master=ventana)  # A tk.DrawingArea.
        canvas.draw()
        canvas.get_tk_widget().place(x=400, y=690, width= 400, height= 230)
        
        toolbar = NavigationToolbar2Tk(canvas, root)
        toolbar.update()
        canvas.get_tk_widget().place(x=540, y=490, width= 150, height= 150)

    if eleccionGriega.get() == "Rho" and opción == "call":
        fig = Figure(figsize=(0.4, 0.4), dpi=100)
        t = np.arange(0, 3, .01)
        fig.add_subplot(111).plot(df["Rho Call"])

        canvas = FigureCanvasTkAgg(fig, master=ventana)  # A tk.DrawingArea.
        canvas.draw()
        canvas.get_tk_widget().place(x=400, y=690, width= 400, height= 230)
        
        toolbar = NavigationToolbar2Tk(canvas, root)
        toolbar.update()
        canvas.get_tk_widget().place(x=540, y=490, width= 150, height= 150)

    if eleccionGriega.get() == "Rho" and opción == "put":
        fig = Figure(figsize=(0.4, 0.4), dpi=100)
        t = np.arange(0, 3, .01)
        fig.add_subplot(111).plot(df["Rho Put"])

        canvas = FigureCanvasTkAgg(fig, master=ventana)  # A tk.DrawingArea.
        canvas.draw()
        canvas.get_tk_widget().place(x=400, y=690, width= 400, height= 230)
        
        toolbar = NavigationToolbar2Tk(canvas, root)
        toolbar.update()
        canvas.get_tk_widget().place(x=540, y=490, width= 150, height= 150)
        
    imagen = PhotoImage(file="graficaD.png")
    labelImagen=tk.Label(ventana, image = imagen)
    labelImagen.place(x=540, y=490, width= 150, height= 150)
   
    
btn_calcular = tk.Button(ventana, text= "Calcular", command= lambda: black_scholes(), background="#4391d7")
btn_calcular.place(x=880,y=280)

labelRespuesta = tk.Label(ventana, text="___________", font=("Arial Black",30), background="#4391d7")
labelRespuesta.place(x=700, y=515)

ventana.mainloop()
```
