##Correlaci칩n/cointegraci칩n

```python
#librerias
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
#importar la bdd
df = pd.read_excel('Matriz RF.xlsx')
```

```python
df = df.iloc[:, 1:]
df
```

Correlaci칩n

```python
#correlaci칩n
matriz_correlacion = df.corr()
print(matriz_correlacion)
```

Cointegraci칩n

```python
#cointegraci칩n

#M0
matriz_cointegracion = np.zeros((df.shape[1], df.shape[1]))

#Iteraci칩n para generar la matriz de CI
for i in range(df.shape[1]):
    for j in range(df.shape[1]):
        if i != j:
            #Prueba CI
            resultado = sm.tsa.stattools.coint(df.iloc[:, i], df.iloc[:, j])
            # P-Value (0.05)
            if resultado[1] < 0.05:
                matriz_cointegracion[i, j] = 1

#df M CI
matriz_cointegracion_df = pd.DataFrame(matriz_cointegracion, columns=df.columns, index=df.columns)
print(matriz_cointegracion_df)
```

Matriz de largos

```python
#matriz largos
matriz_resultante = np.zeros((43, 43))

#Iterar sobre las matrices de cointegraci칩n y  correlaci칩n
for i in range(43):
    for j in range(43):
        if matriz_cointegracion_df.iloc[i, j] == 1 and matriz_correlacion.iloc[i, j] >= 0.70:
            matriz_resultante[i, j] = 1

# Convertir la matriz en un DataFrame de pandas para una mejor visualizaci칩n
matriz_largos = pd.DataFrame(matriz_resultante, columns=matriz_cointegracion_df.columns, index=matriz_cointegracion_df.index)

# Mostrar la matriz resultante
print(matriz_largos)
```

Ubicaci칩n (nombres) matriz de largos

```python
#nombres largo
# Iterar sobre las filas y columnas de la matriz resultante
for i in range(43):
    for j in range(43):
        # Verificar si el valor es 1
        if matriz_resultante[i, j] == 1:
            # Obtener el nombre de la fila y la columna
            nombre_fila = df.index[i]
            nombre_columna = df.columns[j]
            print(f'Hay un 1 en la fila "{nombre_fila}", columna "{nombre_columna}"')
```

Matriz de cortos

```python
#matriz cortos
matriz_resultante = np.zeros((43, 43))

#Iterar sobre las matrices de cointegraci칩n y  correlaci칩n
for i in range(43):
    for j in range(43):
        if matriz_cointegracion_df.iloc[i, j] == 1 and matriz_correlacion.iloc[i, j] <= -0.70:
            matriz_resultante[i, j] = 1

# Convertir la matriz en un DataFrame de pandas para una mejor visualizaci칩n
matriz_largos = pd.DataFrame(matriz_resultante, columns=matriz_cointegracion_df.columns, index=matriz_cointegracion_df.index)

# Mostrar la matriz resultante
print(matriz_largos)
```

Nombres (ubicaci칩n) cortos

```python
#nombres corto
# Iterar sobre las filas y columnas de la matriz resultante
for i in range(43):
    for j in range(43):
        # Verificar si el valor es 1
        if matriz_resultante[i, j] == 1:
            # Obtener el nombre de la fila y la columna
            nombre_fila = df.index[i]
            nombre_columna = df.columns[j]
            print(f'Hay un 1 en la fila "{nombre_fila}", columna "{nombre_columna}"')
```
